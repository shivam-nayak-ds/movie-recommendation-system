```python
def convert_cast(obj):
    L = []
    counter = 0
    for i in ast.literal_eval(obj):
        if counter < 3:
            L.append(i['name'])
            counter += 1
        else:
            break
    return L
```

```python
def fetch_director(obj):
    L = []
    for i in ast.literal_eval(obj):
        if i['job'] == 'Director':
            L.append(i['name'])
            break
    return L
```

# 🎬 Movie Recommendation System

## 📌 Project Overview
A Content-Based Filtering system that recommends similar movies
based on genres, keywords, cast, and director using the
TMDB 5000 Movies Dataset.

## 📂 Dataset
- `tmdb_5000_movies.csv` — Contains movie details such as
   genres, keywords, and overview
- `tmdb_5000_credits.csv` — Contains cast and crew information

## ⚙️ Workflow

### Step 1 — Import Libraries
Imported essential libraries including pandas, numpy,
ast, scikit-learn, and pickle.

### Step 2 — Data Loading
- Loaded both CSV files into pandas DataFrames
- Verified shapes: Movies (4803, 20) | Credits (4803, 4)
- Merged both DataFrames on movie ID

### Step 3 — Data Preprocessing
- Selected relevant columns: id, title, overview,
  genres, keywords, cast, crew
- Parsed JSON formatted columns to extract names
- Extracted top 3 actors from cast
- Extracted Director from crew
- Handled null values

### Step 4 — Feature Engineering
- Combined overview, genres, keywords, cast, and director
  into a single `tags` column
- Each movie is now represented as a single text string

## 🔜 Upcoming Steps
- Text Vectorization using CountVectorizer
- Similarity Calculation using Cosine Similarity
- Building Recommendation Function
- Streamlit App Development
- Deployment on Streamlit Cloud

```python
from sklearn.feature_extraction.text import CountVectorizer

cv = CountVectorizer(max_features=5000, stop_words='english')
vectors = cv.fit_transform(new_df['tags']).toarray()

print(vectors.shape)
```

```python
cv.get_feature_names_out()
```

```python
vocab = cv.get_feature_names_out()
print(vocab[:150])
```

```python
import nltk
from nltk.stem.porter import PorterStemmer

nltk.download('punkt')

ps = PorterStemmer()

def stem(text):
    return " ".join([ps.stem(word) for word in text.split()])

new_df['tags'] = new_df['tags'].apply(stem)

print(new_df['tags'][0])
```

```python
from sklearn.metrics.pairwise import cosine_similarity
similarity = cosine_similarity(vectors)
similarity[0]
```

```python
def recommend(movie):

    index = new_df[new_df['title'] == movie].index[0]


    similar_movies = sorted(list(enumerate(similarity[index])), reverse=True, key=lambda x: x[1])[1:6]

    for i in similar_movies:
        print(new_df.iloc[i[0]].title)
```

```python
recommend('Avatar')
```

```python
recommend('Batman Begins')
```

```python
import pickle

pickle.dump(new_df, open('movies.pkl', 'wb'))
pickle.dump(similarity, open('similarity.pkl', 'wb'))

print("Files saved!")
```

## 🤖 Model Building

### Text Vectorization
- Converted movie tags into numbers using CountVectorizer
- Used top 5000 words only
- Removed common words like "the", "is", "and"
- Applied Stemming — "loving" becomes "love"

### Cosine Similarity
- Compared all 4803 movies with each other
- Higher score means more similar movies

### Recommendation Function
- Input: Movie name
- Output: Top 5 similar movies
